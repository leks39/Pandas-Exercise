```python
##Filtering DataFrames ##
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[movies.duration >= 200] #filtering for movies using 1 condition (duration over 200mins)
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]  #filtering movies multiple conditions & used instead of and 
```

```python
movies[(movies.duration >= 200) | (movies.genre == 'Drama')]  #filtering movies multiple conditions | used instead of or 
```

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Adventure'])] #filtering using multiple OR conditions
```

```python
##Group by Pandas##
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.beer_servings.mean() ##mean of bear_servings
```

```python
continent_drink = drinks.groupby('continent').beer_servings.mean() #Q: which continent drinks most beers
```

```python
continent_drink.sort_values(ascending = False)
```

```python
continent_drink.max()
```

```python
drinks.groupby('continent').wine_servings.describe() #Q: Summarize the statistics of wine serving by continent
```

```python
continent_alcohol = drinks.groupby('continent').mean() #Q: average alcohol consumption by continent
```

```python
continent_alcohol
```

```python
drinks.groupby('continent').spirit_servings.agg(['mean','median','max']) #Q: find the mean,median,max of spirit servings
```

```python
drinks[drinks.continent=='Africa'] #filtering by categories i.e only African countries
```

```python
drinks.groupby('continent').beer_servings.agg(['count', 'mean', 'max','min']) #agg allows for inclusions of multiple stats
```

```python
import pandas as pd
```

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

```python
df = pd.DataFrame(raw_data)
```

```python
df
```

```python
#find mean preTest score for Nighthawks regiment
NGH = df.groupby('regiment').get_group('Nighthawks') #getting details for a particular regiment using groupby
```

```python
NGH.preTestScore.mean()
```

```python
#find general statistics by company
Company_Stats = df.groupby('company').describe()
```

```python
Company_Stats
```

```python
#mean of each company's pretest score
Company_Mean = df.groupby('company').preTestScore.mean()
```

```python
Company_Mean
```

```python
#find pretest score mean grouped by regiment and company
 df.groupby(['company', 'regiment']).preTestScore.mean()
```

```python
#find pretest score mean grouped by regiment and company
df.groupby(['regiment', 'company']).preTestScore.mean() #multiple groupby is inoutted in form of a list
```

```python
#group the entire dataframe by regiment and company
new_df = df.groupby(['regiment', 'company'])
```
