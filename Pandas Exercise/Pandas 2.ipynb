```python
import pandas as pd
```

```python
users = pd.read_table('http://bit.ly/movieusers')
users.head()
```

```python
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zipcode']
users2 = pd.read_table('http://bit.ly/movieusers', sep='|', header = None, names = user_cols, index_col = 'user_id')
```

```python
users2.head() #instructions like sep, header, names, index_col cleans, makes user2 table better than user table
```

```python
users2.shape #shape identifies no of rows and columns
```

```python
users2.duplicated() #checking for duplicate values in the users2 df
```

```python
users2.duplicated().sum() #counting no of duplicated values
```

```python
users2.loc[users2.duplicated(), :] #gets details of the duplicated rows
```

```python
users2.drop_duplicates(keep = 'first') #no of rows changes from 943 to 936 as seen in shape, keep = first keeps initial set
```

```python
users2.drop_duplicates(keep = 'last') #no of rows changes also from 943 to 936 keep = last keeps last set of duplicates
```

```python
users2.drop_duplicates(keep = False) #no of rows changes also from 943 to 929 because all duplicates have been deleted
```

```python
users2.duplicated(subset = ['age', 'zipcode']).sum() #dropping duplicates considering certain columns
```

```python
users2.drop_duplicates(subset = ['age', 'zipcode'])
```

Adding, Dropping Columns using inplace

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.shape #check nos of rows and columns
```

```python
ufo.head()
```

```python
ufo.drop('City', axis =1).head() #removing the City column. axis= 1 is selected because it is a column
```

```python
ufo.drop('City', axis =1, inplace = True) #inplace = True makes changes permanent
```

```python
ufo.head()
```

```python
ufo.set_index('Time', inplace = True) #set index makes a particular column the reference point of the df
```
